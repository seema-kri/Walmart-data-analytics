```python
customer=pd.read_csv("clean_customer.csv")
product = pd.read_csv("clean_product.csv")
store = pd.read_csv("clean_store.csv")
employee = pd.read_csv("clean_employee.csv")
sales = pd.read_csv("clean_sales.csv")

product.to_sql('clean_product', engine, if_exists='replace', index=False)
store.to_sql('clean_store', engine, if_exists='replace', index=False)
employee.to_sql('clean_employee', engine, if_exists='replace', index=False)
sales.to_sql('clean_sales', engine, if_exists='replace', index=False)
customer.to_sql('clean_customer', engine, if_exists='replace', index=False)
```
